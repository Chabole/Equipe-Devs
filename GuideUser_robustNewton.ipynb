```python
import numpy as np
import ZebraLib as zb
import pandas as pd
import matplotlib.pyplot as plt
import random
from tqdm import tqdm
```

```python
x = np.linspace(0, 10)

plt.plot(x, np.cos(x*0.8)*1, label=r'$cos(x)$')
plt.plot(x, np.sin(x)*0.5, label=r'$sin(x)$')
plt.legend()
```

```python
def raiz_separadas(x, args):
    x1, x2 = x
    c1, c2 = args
    return np.array(([np.cos(x1*0.8)*c1,
                     np.sin(x2)*c2]))

def resid_intersec(x, args):
    c1, c2 = args
    return np.cos(x*0.8)*c1 - np.sin(x)*c2

raiz_sep = zb.robustNewton(raiz_separadas, (4., 2), args=(1., 0.5))
inter = zb.robustNewton(resid_intersec, (4.), args=(1., 0.5))

print(f'Raizes inidividual de cada função cos: {raiz_sep[0][0]:.2f} e sin: {raiz_sep[0][1]:.2f}')
print(f'Interseção funções {inter[0]:.2f}')
```
